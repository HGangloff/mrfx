```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

cpu_device = jax.devices('cpu')[0]
```

```python
key = jax.random.PRNGKey(0)
```

```python
from mrfx.models import GUM
from mrfx.samplers import GUMSampler
```

```python
kappa = 0.1
```

```python
gum = GUM(kappa=kappa, K=6, dim=2)
gum_sampler = GUMSampler(n_bands=2500, lx=1024, ly=1024, method="spectral")
key, subkey = jax.random.split(key, 2)
gum_x = gum_sampler.sample_image(gum, subkey)[0]
plt.imshow(gum_x)
plt.show()
```
